```python
for i in file:
    if i.endswith('angry.wav'):
        response.append('angry')
    
    if i.endswith('disgust.wav'):
        response.append('disgust')
    
    if i.endswith('fear.wav'):
        response.append('fear')
    
    if i.endswith('happy.wav'):
        response.append('happy')
    if i.endswith('neutral.wav'):
        response.append('neutral')
    if i.endswith('ps.wav'):
        response.append('Pleasently Suprised')
    if i.endswith('sad.wav'):
        response.append('Sad')
```
